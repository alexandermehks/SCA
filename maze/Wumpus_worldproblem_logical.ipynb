# moveAgent
Implemented a node search where I track on which node the agent is at currently, then im searching all the nodes around that node to make sure that there is no danger on these nodes. After the agent has moved to a new node im adding that node to the visited list. I need to do this so the agent wont visit the same node more than once. In the case that agent gets stuck with a visited node behind it and Wumpus/Pit ahead of it, it will reset, but It will keep the visited list of nodes where the path led to the agent getting stuck, to prevent that in the next run.

# solvable
I'm conducting a version of the Breadth first search or BFS. I'm firstly initiating a list that will represent what will be a invalid search node, in this case it's the Wumpus or an pit. The reason Im not including Wumpus node and nodes where there is a pit is because I want to be able to search the maze for possibilites to get to the gold without stepping into a pit or wumpus, if I would include neighbors of pit or wumpus I could get a hit on the gold without there being a way for the agent to enter and get the gold, and that would be an unsolvable maze. The solvable method is actually being run as recursion where im using the neighbors of the start node as arguments for the next execution until there is no nodes left. 

In this way im able to get the neighbors of all the walkable nodes, and to trigger if the activated node is having gold or glitter in it, and if so it will be solvable.
